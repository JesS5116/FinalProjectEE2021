```python
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
import numpy as np
```

```python
ruoa       = pd.read_csv('../002_datos/ruoafile.csv',index_col=0,parse_dates=True)
split_ruoa = ruoa.truncate(before='2019-05-03 00:00:00',after='2019-06-25 00:00:00')
split_ruoa.drop(['To','hr','Igb','Pr'],axis='columns',inplace=True)
split_ruoa.to_csv('../002_datos/truncateruoa.csv')
#split_ruoa
```

```python
esolmet  = pd.read_csv('../002_datos/datalimpia.csv',index_col=0,parse_dates=True)
# Drop columns
esolmet.drop(['UV','Wspeed'],axis='columns',inplace=True)
#esolmet
```

```python
datos = pd.concat([esolmet,split_ruoa],axis=1)
datos
```

```python
datos.to_csv('../002_datos/concatEsolmetRuoa.csv')
```

```python
data = pd.read_csv('../002_datos/concatEsolmetRuoa.csv',index_col=0,parse_dates=True)
data
```

```python
# Split year. month, day, minute, second
data['year']=data.index.year
data['month']=data.index.month
data['day']=data.index.day
data['hour']=data.index.hour
data['minute']=data.index.minute

# Rearange the columns
datac = data.copy()
datac = datac[['year','month','day','hour','minute','Idirecta','Iglobal','Idifusa',
             'To','HR','ws','wd','P']]
```

```python
datac.to_csv('../002_datos/esolmetruoasplit.csv')
```
