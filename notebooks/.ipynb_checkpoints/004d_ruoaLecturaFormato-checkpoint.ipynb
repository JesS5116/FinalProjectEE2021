```python
import pandas as pd
import numpy as np
```

```python
# Call file
ruoa = pd.read_csv('../002_datos/ruoa.txt',skiprows=4,delimiter="\t",usecols=[0,1,2,3,4,8,9],
                   names=['tiempo','To','hr','ws','wd','P','Ig'],index_col=0,parse_dates=True)
# Set index
ruoa.sort_index(inplace=True)
# Resample with 6 timestep
ruoa = ruoa.resample("600S").interpolate(method='time')
ruoa
```

```python
# Nuevas columnas en Pascales y con radiación cero
ruoa['Pr']=ruoa.P*100
ruoa['Igb']=abs(ruoa.Ig*0)
# Borrar columna de presión en hPa y Ig
ruoa.drop('P',axis='columns',inplace=True)
ruoa.drop('Ig',axis='columns',inplace=True)
```
