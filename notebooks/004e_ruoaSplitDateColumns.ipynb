```python
import pandas as pd
import numpy as np
```

```python
# Call file
ruoa = pd.read_csv('../002_datos/ruoafile.csv',index_col=0,parse_dates=True)
ruoa
```

```python
# Split year. month, day, minute, second
ruoa['year']=ruoa.index.year
ruoa['month']=ruoa.index.month
ruoa['day']=ruoa.index.day
ruoa['hour']=ruoa.index.hour
ruoa['minute']=ruoa.index.minute

# Rearange the columns
data = ruoa.copy()
data = data[['year','month','day','hour','minute','To','hr','ws','wd','Pr','Igb']]
```
