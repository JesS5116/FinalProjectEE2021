```python
import pandas as pd
import numpy as np
```

```python
# Call file
ruoa = pd.read_csv('../002_datos/ruoa.txt',skiprows=4,delimiter="\t",usecols=[0,1,2,3,4,8,9],
                   names=['tiempo','To','hr','ws','wd','P','Ig'],index_col=0,parse_dates=True)
# Set index
ruoa.sort_index(inplace=True)
# Resample with 6 timestep
ruoa = ruoa.resample("600S").interpolate(method='time')
ruoa
```

```python
# Change every value of the Radiation to 0
for i in np.arange(0,48960,1):
    ruoa.Ig.replace(to_replace=ruoa.Ig[i],value=0,inplace=True)
    
ruoa
```

```python
ruoa.to_csv('../002_datos/ruoafile.csv')
```
